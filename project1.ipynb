```python
import pandas as pd
df = pd.read_csv('orders.csv')
df.head(20)
```

```python
df.fillna(0, inplace=True)          # Replace missing values with 0
df.dropna(inplace=True)             # Drop rows with any missing values
print(df)
```

```python
df.dropna(axis=1, inplace=True)             # Drop columns with any missing values
print(df)
```

```python
df.rename(columns={
    'Order Id': 'order_id',
    'Order Date': 'order_date',
    'Ship Mode': 'ship_mode',
    'Postal Code': 'postal_code',
    'Sub Category': 'sub_category',
    'Product Id': 'product_id',
    'cost price': 'cost_price',
    'List Price': 'list_price',
    'Discount Percent': 'discount_percent'
}, inplace=True)
df.head(2)
```

```python
df = df.map(lambda x: x.rstrip() if isinstance(x, str) else x)          # Apply rstrip() to remove trailing spaces from all string columns
```

```python
df['discount'] = df['list_price'] * (df['discount_percent']/100)
df['sale_price'] = df['list_price']-df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
df['Revenue'] = df['sale_price'] * df['Quantity']
print(df.head(2))
```

```python
df.to_csv('orders.csv', index=False)
```

```python
# 1.Find top 10 highest revenue generating products
sql_query = """
SELECT product_id, SUM(quantity * sale_price) AS total_revenue, Category, City
FROM orders
GROUP BY product_id
ORDER BY total_revenue DESC
LIMIT 10;
"""

# Executing the query
mycursor.execute(sql_query)

# Fetching the results
results = mycursor.fetchall()

# Printing the results
for row in results:
    print(f"Product ID: {row[0]}, Total Revenue: {row[1]}, Category: {row[2]}, City: {row[3]}")

# Close the cursor and connection
#mycursor.close()
#mydb.close()
```

```python
# 2.Find the top 5 cities with the highest profit margins
sql_query = """
SELECT City, product_id, SUM(Revenue) AS total_revenue, SUM(cost_price) as total_cost,
(SUM(Revenue) - SUM(cost_price)) / SUM(Revenue) * 100 AS profit_margin
FROM orders
GROUP BY City
ORDER BY profit_margin DESC
LIMIT 5;
"""

# Executing the query
mycursor.execute(sql_query)

# Fetching the results
results = mycursor.fetchall()

# Printing the results
for row in results:
    print(f"City: {row[0]}, Product ID: {row[1]}, Total Revenue: {row[2]}, Total_Cost: {row[3]}, profit Margin:{row[4]}")
```

```python
# 3.Calculate the total discount given for each category

sql_query = """
SELECT Category, SUM(discount) AS total_discount
FROM Orders
group by Category
Order by total_discount  desc;
"""

# Executing the query
mycursor.execute(sql_query)

# Fetching the results
results = mycursor.fetchall()

# Printing the results
for row in results:
    print(f"Category: {row[0]}, Total Discount: {row[1]}")
```

```python
# 4.Find the average sale price per product category

sql_query = """
SELECT Category, AVG(sale_price) AS Avg_sale_price
FROM Orders
group by Category
Order by Avg_sale_price  desc;
"""

# Executing the query
mycursor.execute(sql_query)

# Fetching the results
results = mycursor.fetchall()

# Printing the results
for row in results:
    print(f"Category: {row[0]}, Average Sale Price: {row[1]}")
```

```python
# 5.Find the region with the highest average sale price

sql_query = """
select Region, AVG(sale_price) as Avg_sale_price
FROM Orders
group by Region
Order by Avg_sale_price desc
Limit 1;"""

mycursor.execute(sql_query)

results = mycursor.fetchall()

for row in results:
    print(f"Region: {row[0]}, Average Sale Price: {row[1]}")
```

```python
# 6.Find the total profit per category

sql_query = """
select Category, SUM(profit) as Total_profit
From Orders
group by Category
Order by Total_profit;"""

mycursor.execute(sql_query)

results = mycursor.fetchall()

for row in results:
    print(f"Category: {row[0]}, Total_Profit: {row[1]}")
```

```python
# 7.Identify the top 3 segments with the highest quantity of orders

sql_query ="""
select Segment, SUM(Quantity) as Total_order_Quantity
From Orders
group by Segment
Order by Total_order_Quantity desc
Limit 3;"""

mycursor.execute(sql_query)

results = mycursor.fetchall()

for row in results:
    print(f"Segment: {row[0]}, Total_order_Quantity: {row[1]}")
```

```python
# 8. Determine the average discount percentage given per region

sql_query = """
select Region, AVG(discount_percent) as  Avg_discount_percent
From Orders
group by Region
order by Avg_discount_percent""" 

mycursor.execute(sql_query)

results = mycursor.fetchall()

for row in results:
    print(f"Region: {row[0]}, Avg Discount Percent: {row[1]}")
```

```python
# 9.Find the product category with the highest total profit

sql_query ="""select Category, SUM(Quantity*profit) as Total_profit
From Orders
group by Category
order by Total_profit
limit 1;"""

mycursor.execute(sql_query)

results = mycursor.fetchall()

for row in results:
    print(f"Category: {row[0]}, Total Profit: {row[1]}")
```
